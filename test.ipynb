```python
import matplotlib.pyplot as plt
import numpy as np

# Parameters for the Lissajous curve
A = 1       # Amplitude in x
B = 1       # Amplitude in y
a = 3       # Frequency in x
b = 2       # Frequency in y
delta = np.pi / 2  # Phase shift

t = np.linspace(0, 2 * np.pi, 1000)
x = A * np.sin(a * t + delta)
y = B * np.sin(b * t)

fig, ax = plt.subplots(figsize=(6,6))
ax.plot(x, y, color='#7f3f98', linewidth=2)

ax.set(title='Lissajous Curve', xlabel='X-axis', ylabel='Y-axis')
ax.grid(True)
plt.show()
```
